```python
#Import required libraries
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import cv2
%matplotlib inline
```

# Accessing and Manipulating Image Pixels

```python
cb_image = "./images/checkerboard_18x18.png"
img = cv2.imread(cb_image,0)
plt.imshow(img,cmap='gray')
```

```python
#Accessing the pixels
print(img[0,10])
print(img[10,3])
```

```python
#Modifying the image pixels
cb_image1 = img.copy()
print(cb_image1[0,10])
print(cb_image1[10,3])
cb_image1[0,10] = 0
cb_image1[10,3] = 0
print(cb_image1[0,10])
print(cb_image1[10,3])
```

# Cropping Images

```python
img_nz = cv2.imread("images/New_Zealand_Lake.jpg")
img_nz = img_nz[:,:,::-1]
plt.imshow(img_nz)
```

```python
#cropped = img[start_row:end_row, start_col:end_col]
img_crop = img_nz[100:200,200:300]
plt.imshow(img_crop)
```

# Resizing Image

dst = resize(src,dsize(,dst(,fx(,fy(,interpolation)))))

fx - scale factor along horizontal axis

fy - scale factor along vertical axis

```python
#specify scale factor using fx and fy

resize_img = cv2.resize(img_crop, None, fx=2,fy=2)
plt.imshow(resize_img)
```

```python
#Specify exact output size
width = 100
height = 100
dim = (width, height)
resize_img = cv2.resize(img_crop, dim, cv2.INTER_AREA)
plt.imshow(resize_img)
```

```python
#Resize maintaining aspect ratio

height, width = img_crop.shape[:2]

# Define the new width while maintaining the aspect ratio
new_width = 100
#aspect ratio based on a given width
aspect_ratio = width / float(height)
new_height = int(new_width / aspect_ratio) 
#resize based on the height
#new_height = 600
#new_width = int(new_height * aspect_ratio)

img_resize = cv2.resize(img_crop, (new_width,new_height),cv2.INTER_AREA)
plt.imshow(img_resize)
```

# Flipping image

cv2.flip(src,flipCode)

flipCode:

0 - flip around x-axis

1 - flip around y-axis

-1 - flip around both axis

```python
img_nz = cv2.imread('images/New_Zealand_Lake.jpg')
img_nz = img_nz[:,:,::-1]
img_flip_x = cv2.flip(img_nz,0)
img_flip_y = cv2.flip(img_nz,1)
img_flip_z = cv2.flip(img_nz,-1)
plt.figure(figsize=(20,5))
plt.subplot(141);plt.imshow(img_nz)
plt.subplot(142);plt.imshow(img_flip_x)
plt.subplot(143);plt.imshow(img_flip_y)
plt.subplot(144);plt.imshow(img_flip_z)
```
