```python
#import necessary libraries

import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
```

```python
#draw line
# cv2.line(src, pt1,pt2, color,thickness,linetype,shift)

img = cv2.imread("images/coca-cola-logo.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.line(img,(200,100),(400,100),(0,0,0),1,cv2.LINE_AA)
plt.imshow(img)
```

```python
#Draw circle
# cv2.circle(src, pt1, radius, color, thickness,linetype,shift)
img = cv2.imread("images/coca-cola-logo.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.circle(img,(200,100),100,(0,0,0),1,cv2.LINE_AA)
plt.imshow(img)
```

```python
#Draw rectangle
# cv2.rectangle(src, pt1,pt2, color,thickness,linetype,shift)

img = cv2.imread("images/coca-cola-logo.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.rectangle(img,(200,100),(600,500),(0,0,0),1,cv2.LINE_AA)
plt.imshow(img)
```

```python
#Add text
#cv2.puText(img, text,org, fontface, fontscale, color, thickness, linetype)

img = cv2.imread("images/coca-cola-logo.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.putText(img,"hello",(0,100),1,2,(0,0,0),cv2.FONT_HERSHEY_PLAIN)
plt.imshow(img)
```
